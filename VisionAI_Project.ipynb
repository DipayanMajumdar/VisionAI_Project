```python
# ============================================================
# 1. Install & Import
# ============================================================
!pip install tensorflow matplotlib scikit-learn seaborn --quiet

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import layers, models
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import os

# ============================================================
# 2. Mount Google Drive
# ============================================================
from google.colab import drive
drive.mount('/content/drive')
save_path = '/content/drive/MyDrive/VisionAI_Project/'
os.makedirs(save_path, exist_ok=True)

# ============================================================
# 3. Load & Preprocess CIFAR-10 Dataset
# ============================================================
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

val_size = 5000
x_val, y_val = x_train[:val_size], y_train[:val_size]
x_train, y_train = x_train[val_size:], y_train[val_size:]

class_names = ['Airplane', 'Automobile', 'Bird', 'Cat', 'Deer',
               'Dog', 'Frog', 'Horse', 'Ship', 'Truck']

plt.figure(figsize=(8,4))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(x_train[i])
    plt.title(class_names[int(y_train[i])])
    plt.axis('off')
plt.show()

# ============================================================
# 4. Baseline CNN Model
# ============================================================
cnn_model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(128, (3,3), activation='relu'),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

cnn_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history_cnn = cnn_model.fit(x_train, y_train, epochs=10, batch_size=64, validation_data=(x_val, y_val))
cnn_model.save(save_path + "baseline_cnn.keras")

# ============================================================
# 5. MobileNetV2 Transfer Learning with Memory-Safe Resizing
# ============================================================
train_gen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True
)
val_gen = ImageDataGenerator()

train_flow = train_gen.flow(x_train, y_train, batch_size=32)
val_flow = val_gen.flow(x_val, y_val, batch_size=32)

resize_layer = layers.Resizing(160, 160)
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(160,160,3))
base_model.trainable = False

inputs = layers.Input(shape=(32, 32, 3))
x = resize_layer(inputs)
x = base_model(x, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(256, activation='relu')(x)
outputs = layers.Dense(10, activation='softmax')(x)

tl_model = models.Model(inputs, outputs)

tl_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history_tl = tl_model.fit(train_flow, validation_data=val_flow, epochs=5)

base_model.trainable = True
for layer in base_model.layers[:-30]:
    layer.trainable = False

tl_model.compile(optimizer=tf.keras.optimizers.Adam(1e-5),
                 loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history_ft = tl_model.fit(train_flow, validation_data=val_flow, epochs=5)

tl_model.save(save_path + "mobilenetv2_finetuned.keras")

# ============================================================
# 6. Evaluation & Plots
# ============================================================
plt.figure(figsize=(8,5))
plt.plot(history_cnn.history['accuracy'], label='CNN Train')
plt.plot(history_cnn.history['val_accuracy'], label='CNN Val')
plt.plot(history_tl.history['accuracy']+history_ft.history['accuracy'], label='MobileNetV2 Train')
plt.plot(history_tl.history['val_accuracy']+history_ft.history['val_accuracy'], label='MobileNetV2 Val')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.savefig(save_path + "accuracy_curves.png")
plt.show()

test_gen = ImageDataGenerator().flow(x_test, y_test, batch_size=32, shuffle=False)
y_pred = np.argmax(tl_model.predict(test_gen), axis=1)

cm = confusion_matrix(y_test.flatten(), y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.ylabel('True')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.savefig(save_path + "confusion_matrix.png")
plt.show()

print(classification_report(y_test.flatten(), y_pred, target_names=class_names))

plt.figure(figsize=(10,5))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(x_test[i])
    pred_label = class_names[y_pred[i]]
    true_label = class_names[int(y_test[i])]
    plt.title(f"P:{pred_label}\nT:{true_label}", fontsize=8)
    plt.axis('off')
plt.tight_layout()
plt.savefig(save_path + "sample_predictions.png")
plt.show()

# ============================================================
# 7. Zip all results for easy download
# ============================================================
!zip -r /content/drive/MyDrive/VisionAI_Project/VisionAI_Project.zip /content/drive/MyDrive/VisionAI_Project/

```
